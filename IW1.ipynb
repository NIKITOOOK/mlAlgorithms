##1.Прочитать данные и создать столбец "Цена со скидкой", вычислив цену товара с учетом скидки.

## Читаем файл product_sheet.csv

```python
pd.read_csv("product_sheet.csv")
```

```python
data = pd.read_csv('product_sheet.csv')
```

```python
print(f'Всего {len(data)} продуктов в выборке')
data.head()
```

#Создать столбец "Цена со скидкой" и вычисляем цену со скидкой

```python
data['Цена со скидкой'] = data['Цена'] * (1 - data['Скидка в процентах']/100)
data.head()
```

# 2.Отсортировать товары по убыванию цены со скидкой и вывод топ-10 самых дорогих товаров

```python
data_sorted = data.sort_values(by='Цена со скидкой', ascending=False)
```

```python
data_sorted.head(10)
```

# 3.Сгруппировать данные по категориям и вычислить среднюю цену товара (как с учетом скидки, так и без неё) по каждой категории.

```python
grouped = data_sorted.groupby('Категория')
grouped
```

```python

grouped = data_sorted.groupby(['Категория'])
print(grouped)
```

```python
mean_salary = data_sorted.groupby('Категория')['Цена'].mean()
mean_salary
```

```python
# Применение различных функций к столбцам
aggregated = data.groupby('Категория').agg({
    'Цена': 'mean',  # Средняя
    'Цена со скидкой': 'mean'
})
aggregated
```
